```python
# year
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(df_cat["year"].values)
cat1 = le.transform(df_cat["year"].values)
df_x["year"] = cat1

#type
le = preprocessing.LabelEncoder()
le.fit(df_cat["type"].values)
cat2 = le.transform(df_cat["type"].values)
df_x["type"] = cat2

# region

le = preprocessing.LabelEncoder()
le.fit(df_cat["region"].values)
cat3 = le.transform(df_cat["region"].values)
df_x["region"] = cat3
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression(normalize=True)
lr.fit(x_train,y_train)
lr_pred=lr.predict(x_test)
```

```python
lr_pred
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, lr_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, lr_pred)
```


squaredbool, default=True If True returns MSE value, if False returns RMSE value.

```python

print("RMSE VALUE = ",mean_squared_error(y_test, lr_pred,squared = False))
```
